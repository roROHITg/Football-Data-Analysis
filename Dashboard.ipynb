```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from PIL import Image

hteam_logo= 'D:\Python Basics\Football Data Analytics\Sofascore\Teams Logo\eb.png'
ateam_logo= 'D:\Python Basics\Football Data Analytics\Sofascore\Teams Logo\mb.png'

hteam= 'East Bengal FC'
ateam= 'Mohun Bagan SG'
score= '1 - 3'

# List of plot image filenames
plot_filenames = ['Home Positions.png', 'Stats.png', 'Away Positions.png', 'Home Shotmap.png', 'momentum.png', 'Away Shotmap.png']
# List of logo image filenames
logo_filenames = [hteam_logo, ateam_logo]

# Load the images
plots = [Image.open(fname) for fname in plot_filenames]
logos = [Image.open(fname) for fname in logo_filenames]

# Resize the logos
logo_size = (150, 150)  # Set desired logo size (width, height)
logos = [logo.resize(logo_size, Image.Resampling.LANCZOS) for logo in logos]

# Create a figure
fig = plt.figure(figsize=(20, 14))
gs = GridSpec(2, 3, figure=fig, wspace=0.05, hspace=0.05)  # Reduce wspace and hspace

# Adjust layout to add space at the top
plt.subplots_adjust(top=0.83)

# Add plots to the figure
for i, plot in enumerate(plots):
    ax = fig.add_subplot(gs[i])
    ax.imshow(plot)
    ax.axis('off')

# Add team logos to the figure
fig.figimage(logos[0], 500, 950, zorder=10)  # Adjust position (x, y) and zorder as needed
fig.figimage(logos[1], 910, 950, zorder=10)  # Adjust position (x, y) and zorder as needed


# Add text to the figure
fig.text(0.51, 0.90, hteam +' VS '+ ateam, ha='center', fontsize=24, weight='bold')
fig.text(0.51, 0.85, score, ha='center', va= 'top', fontsize=30, weight= 'bold')


# add credits
CREDIT = "Twitter: @Rohit_speaking"

fig.text(
    0.90, 0.1, f"{CREDIT}", size=15,
    color="#000000",
    ha="right"
)

# Save the final image
plt.savefig('combined_plot.png',  bbox_inches='tight')
plt.show()
```
